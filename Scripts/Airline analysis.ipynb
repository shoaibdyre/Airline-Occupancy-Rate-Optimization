### Basic Analysis

```python
# How many planes have more than 100 seats
pd.read_sql_query("""select aircraft_code, count(*) as num_seats from seats group by aircraft_code having num_seats > 100 """, conn)
```

```python
# How the number of tickets booked and total amount earend changed with time
tickets = pd.read_sql_query(""" select * from tickets inner join bookings on tickets.book_ref = bookings.book_ref """, conn)
tickets['book_date'] = pd.to_datetime(tickets['book_date'])
tickets['date'] = tickets['book_date'].dt.date
x = tickets.groupby('date')[['date']].count()
plt.figure(figsize = (18,6))
plt.plot(x.index, x['date'], marker = '^')
plt.xlabel('Date' , fontsize = 20)
plt.ylabel('Number of Tickets', fontsize = 20)
plt.grid('b')
plt.show()
```

```python
bookings = pd.read_sql_query( """ select * from  bookings """, conn)
bookings['book_date'] = pd.to_datetime(bookings['book_date'])
bookings['date'] = bookings['book_date'].dt.date
x = bookings.groupby('date')[['total_amount']].sum()
plt.figure(figsize = (18,6))
plt.plot(x.index, x['total_amount'], marker = '^')
plt.xlabel('Date' , fontsize = 20)
plt.ylabel('Total amount Earned', fontsize = 20)
plt.grid('b')
plt.show()
```

```python
df = pd.read_sql_query(""" select fare_conditions, aircraft_code, avg(amount) 
                       from ticket_flights
                        join flights on ticket_flights.flight_id = flights.flight_id 
                       group by  aircraft_code, fare_conditions""", conn)

df
```

```python
sns.barplot(data = df, x= 'aircraft_code', y = 'avg(amount)', hue = 'fare_conditions')
```

### Occupancy Rate

```python
# For each Aircraft, calculate the total revencue per year and the average revenue per ticket
pd.read_sql_query("""select aircraft_code, ticket_count, total_revenue, total_revenue/ticket_count as  from
                  (select aircraft_code, count(*) as ticket_count, sum(amount) as total_revenue, 
                  from ticket_flights
                  join flights
                  on ticket_flights.flight_id = flights.flight_id
                  group by aircraft_code) """, conn)
```
